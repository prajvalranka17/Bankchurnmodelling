# BANK CHURN MODELLING

# Aim :
The goal of the project was to develop an ANN(Artificial Neural Network) model using python 
which helped predict churning of bank customers based on past churn dataand other independent variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset
```

```python
X = dataset.iloc[:,3:13]
y = dataset.iloc[:,13]
```

```python
geography = pd.get_dummies(X["Geography"],drop_first=True)
gender = pd.get_dummies(X["Gender"],drop_first=True)
```

```python
X = pd.concat([X,geography,gender],axis=1)
```

```python
X = X.drop(["Geography","Gender"],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc  = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation = 'relu',input_dim = 11))
ann.add(Dropout(0.3))
```

```python
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(0.4))
```

```python
ann.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
ann.add(Dropout(0.2))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann_train = ann.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
y_preds = ann.predict(X_test)
y_preds = (y_preds>0.5)
y_preds
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_preds)
cm
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_preds)
score
```
